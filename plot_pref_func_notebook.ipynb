```python
import numpy as np
import matplotlib.pyplot as plt

# Define the preference functions
def usual_preference(x, threshold=0):
    return np.where(x > threshold, 1, 0)

def linear_preference(x, lower_threshold=0, upper_threshold=1):
    return np.clip((x - lower_threshold) / (upper_threshold - lower_threshold), 0, 1)

def gaussian_preference(x, mean=0, std_dev=1):
    return 1 - np.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))

# Define the range of x values
x_values = np.linspace(start=0, stop=1, num=10000)

# Calculate the preference values
usual_pref_values = usual_preference(x_values)
linear_pref_values = linear_preference(x_values, 0.2, 0.8)
gaussian_pref_values = gaussian_preference(x_values, 0, 0.32)

# Plot the preference functions
plt.figure(figsize=(10, 6))
#plt.axvline(x=0, color='blue', linestyle='--', label='Usual Preference Function')


plt.plot(x_values, usual_pref_values, label='Usual Preference Function', linestyle='--')
plt.plot(x_values, linear_pref_values, label='Linear Preference Function', linestyle='-.')
plt.plot(x_values, gaussian_pref_values, label='Gaussian Preference Function', linestyle='-')

plt.title('Preference Functions')
plt.xlabel('Criterion Difference')
plt.ylabel('Preference')
plt.legend()
plt.grid(True)
plt.show()
```
